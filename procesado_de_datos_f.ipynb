```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./diabetes_012_health_indicators_BRFSS2015.csv")
```

```python
diabetes = df[ df['Diabetes_012'] == 2 ]
sin_diabetes = df[ df['Diabetes_012'] == 0 ]
prediabetes = df[ df['Diabetes_012'] == 1 ]
```

```python
correlacion = df.corr()
numero_de_columnas = len(correlacion)
correlacion.style.background_gradient(cmap='coolwarn')

x = range(numero_de_columnas)
y = correlacion.columns

plt.imshow(correlacion, cmap='hot', interpolation='none')
plt.colorbar()
plt.xticks(x, y, rotation='vertical')
plt.yticks(x, y)
plt.gcf().set_size_inches(10, 10)
```

```python
x = range(len(diabetes))
y = diabetes.columns[1:]

for col in y:
    plt.title(col)
    plt.boxplot([diabetes[col], prediabetes[col], sin_diabetes[col]], labels=['diabetes', 'prediabetes', 'sin_diabetes'])
    plt.show()
```
