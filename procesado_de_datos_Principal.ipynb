```python
df = pd.read_csv("./diabetes_012_health_indicators_BRFSS2015.csv")
df2 = pd.read_csv("./diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
```

```python
diabetes = df[ df['Diabetes_012'] == 2 ]
sin_diabetes = df[ df['Diabetes_012'] == 0 ]
prediabetes = df[ df['Diabetes_012'] == 1 ]

diabetes2 = df2[ df2['Diabetes_binary'] == 1 ]
sin_diabetes2 = df2[ df2['Diabetes_binary'] == 0 ]
```

```python
fig, axs = plt.subplots(2, 3, figsize=(8, 6))

columnas = ['CholCheck','HeartDiseaseorAttack','PhysActivity','HvyAlcoholConsump','AnyHealthcare','DiffWalk',]
columnas2 =  ['HighBP','Smoker']
for i, col in enumerate(columnas2):
    
    counts = diabetes[col].value_counts() 
    axs[i, 0].bar(counts.index, counts.values, color=['blue', 'red']) 
    axs[i, 0].set_title(f'Diabetes {col}', fontsize=8) 
    axs[i, 0].set_xticks([0, 1], ['0', '1'])
   
    counts = prediabetes[col].value_counts() 
    axs[i, 1].bar(counts.index, counts.values, color=['blue', 'red']) 
    axs[i, 1].set_title(f'prediabetes {col}', fontsize=8) 
    axs[i, 1].set_xticks([0, 1], ['0', '1'])

    counts = sin_diabetes[col].value_counts() 
    axs[i, 2].bar(counts.index, counts.values, color=['blue', 'red']) 
    axs[i, 2].set_title(f'sin diabetes {col}', fontsize=8) 
    axs[i, 2].set_xticks([0, 1], ['0', '1'])


plt.tight_layout()
plt.show()
```

"Conclusi√≥n": Se puede analizar que tanto el highBP (hipertensi√≥n) como el highChol (colesterol alto), tanto en las personas que poseen diabetes ,prediabetes y las que no; al tener altos √≠ndices de estos, podr√≠a deberse a que la diabetes, al tener resistencia a la insulina, afecta tambi√©n a otros procesos metab√≥licos, como el control del colesterol y la presi√≥n arterial. Adem√°s, afecta c√≥mo el cuerpo procesa las grasas, lo que puede llevar a niveles elevados de colesterol "malo", ocasionando enfermedades como enfermedad coronaria, neuropat√≠a, accidente cerebrovascular, entre otras; por lo que se puede teorizar que las personas evaluadas no tienen una alimentaci√≥n saludable estricta, ni realizan la actividad f√≠sica necesaria que necesitan para encontrarse saludables y evitar las enfermedades mencionadas.
